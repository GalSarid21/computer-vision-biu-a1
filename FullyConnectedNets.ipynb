## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
AdaGrad accumulates cache values squared gradients from all previous steps, so it only grows over time. this makes the denominator go larger and larger overtime, which shrinks the effective learning rate and causes weights update to become very small. Adam avoids this because it uses an exponentially decaying average of past gradients instead of a cumulative sum, which makes old gradients fade out and the step sizes stay more stable during training.
